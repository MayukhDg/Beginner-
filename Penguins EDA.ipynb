```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
penguins= sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
penguins.isnull().sum()
```

```python
penguins ['species'].value_counts()
```

```python
penguins.dtypes
```

```python
penguins['bill_length_mm'].fillna(penguins['bill_length_mm'].mean(), inplace=True)
```

```python
penguins['bill_depth_mm'].fillna(penguins['bill_depth_mm'].mean(), inplace=True)
penguins['flipper_length_mm'].fillna(penguins['flipper_length_mm'].mean(), inplace=True)
penguins['body_mass_g'].fillna(penguins['body_mass_g'].mean(), inplace=True)
```

```python
penguins.isnull().sum()
```

```python
penguins['sex'].value_counts()
```

```python
penguins['sex'].fillna('Male', inplace=True)
```

```python
sns.barplot(x='species', y='body_mass_g', data=penguins)
```

```python
penguins.columns
```

```python
penguins.rename(columns={'body_mass_g': 'weight', 'bill_depth_mm': 'bill_depth', 'bill_length_mm': 'bill_length', 'flipper_length_mm':'flipper_length' }, inplace=True)
```

```python
penguins.head()
```

```python
plt.figure(figsize=(10,8))
plt.title('Relationship b/w bill length and bill depth')
sns.scatterplot(x='bill_length', y='bill_depth', hue='species', data=penguins)
```

```python
plt.figure(figsize=(10,8))
plt.title('Relationship b/w bill length and bill depth with hue as Sex')
sns.scatterplot(x='bill_length', y='bill_depth', hue='sex', data=penguins)
```

```python
sns.countplot(x='island', hue='sex', data=penguins)
```
